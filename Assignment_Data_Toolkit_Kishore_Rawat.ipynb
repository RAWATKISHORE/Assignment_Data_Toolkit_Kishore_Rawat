##Q3. Explain the following terms:

##* The difference in np.array, np.asarray and np.asanyarray.

##* The difference between Deep copy and shallow copy.

##Ans. Certainly! I'll explain these terms for you.

##Differences between np.array, np.asarray, and np.asanyarray:

##a) np.array:

###* Creates a new NumPy array from input data.
###* Always creates a new copy of the input data.
###*Can specify the data type of the array.

##b) np.asarray:

###* Converts input to a NumPy array.
###* If the input is already a NumPy array with the same dtype, it returns the input without copying.
###*If a copy is needed (different dtype or not a NumPy array), it creates a new array.

##c) np.asanyarray:

###* Similar to np.asarray, but it also preserves subclasses of ndarray.
###* If the input is already a subclass of ndarray, it returns it without copying.
###* If a copy is needed, it creates a new array of the same subclass as the input.

##Key differences:

###* np.array always creates a new array.
###* np.asarray and np.asanyarray may return the input without copying if possible.
###* np.asanyarray preserves subclasses of ndarray, while np.asarray doesn't.



##Difference between Deep copy and Shallow copy:

##a) Shallow Copy:

###* Creates a new object, but references the same memory addresses for nested objects.
###* Changes to nested objects in the copy will affect the original, and vice versa.
###* In NumPy, operations like slicing often create shallow copies.

##b) Deep Copy:

###* Creates a new object and recursively copies all nested objects.
###* Changes to the copy (or its nested objects) don't affect the original, and vice versa.
###* In NumPy, you can use the np.copy() function or the .copy() method for deep copying.

###* Example to illustrate:

```python
import numpy as np

# Shallow copy
a = np.array([[1, 2, 3], [4, 5, 6]])
b = a  # Shallow copy
b[0, 0] = 99
print(a)  # Will show [[99, 2, 3], [4, 5, 6]]

# Deep copy
c = np.array([[1, 2, 3], [4, 5, 6]])
d = c.copy()  # Deep copy
d[0, 0] = 99
print(c)  # Will show [[1, 2, 3], [4, 5, 6]]
```

##Ans. NumPy array with random integers between 1 and 10 of shape (5, 6), and then perform the requested operations.

##Q14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:

###a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').

###b) Label the x-axis as 'Date'.

###c) Set the title of the plot as 'Temperature and Humidity Over Time'.

##Ans.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a time-series dataset
dates = pd.date_range(start='2023-01-01', periods=100)  # 100 days starting from Jan 1, 2023
temperature = np.random.uniform(low=20, high=35, size=100)  # Random temperatures between 20 and 35 degrees Celsius
humidity = np.random.uniform(low=30, high=100, size=100)    # Random humidity between 30% and 100%

# Create a DataFrame
data = pd.DataFrame({
    'Date': dates,
    'Temperature': temperature,
    'Humidity': humidity
})

# Set the 'Date' column as the index
data.set_index('Date', inplace=True)

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 5))

# Step a: Plot 'Temperature' on the left y-axis
color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color=color)
ax1.plot(data.index, data['Temperature'], color=color, label='Temperature')
ax1.tick_params(axis='y', labelcolor=color)

# Step a: Create a second y-axis for 'Humidity'
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Humidity (%)', color=color)
ax2.plot(data.index, data['Humidity'], color=color, label='Humidity')
ax2.tick_params(axis='y', labelcolor=color)

# Step b: Label the x-axis
ax1.set_xlabel('Date')

# Step c: Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.show()
```

##Q15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:

###a) Plot a histogram of the data with 30 bins.

###b) Overlay a line plot representing the normal distribution's probability density function (PDF).

###c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

###d) Set the title of the plot as 'Histogram with PDF Overlay'.

##Ans.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Step 1: Create a NumPy array containing 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # Mean = 0, Standard deviation = 1

# Step 2: Plot a histogram of the data with 30 bins
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Step 3: Overlay a line plot representing the normal distribution's PDF
# Calculate the PDF values for the bins
bin_centers = 0.5 * (bins[1:] + bins[:-1])  # Centers of the bins
pdf = norm.pdf(bin_centers, loc=0, scale=1)  # Calculate the PDF

plt.plot(bin_centers, pdf, 'k', linewidth=2, label='Normal PDF')

# Step 4: Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Step 5: Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

##Q17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

##Ans.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create two random arrays
x = np.random.uniform(-10, 10, size=100)
y = np.random.uniform(-10, 10, size=100)

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y})

# Determine the quadrant for each point
def determine_quadrant(row):
    if row['X'] >= 0 and row['Y'] >= 0:
        return 'Quadrant I'
    elif row['X'] < 0 and row['Y'] >= 0:
        return 'Quadrant II'
    elif row['X'] < 0 and row['Y'] < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

data['Quadrant'] = data.apply(determine_quadrant, axis=1)

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='X', y='Y', hue='Quadrant', palette='viridis', s=100)

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title of the plot
plt.title('Quadrant-wise Scatter Plot')

# Show the legend
plt.legend(title='Quadrants')

# Show the plot
plt.axhline(0, color='gray', linewidth=0.5, linestyle='--')  # Add x-axis
plt.axvline(0, color='gray', linewidth=0.5, linestyle='--')  # Add y-axis
plt.grid()
plt.show()
```

##Q19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.

##Ans.

```python
import random
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import RdYlBu11

# Generate random data
categories = ['A', 'B', 'C', 'D', 'E']
values = [random.randint(1, 100) for _ in categories]

# Create a color mapper
color_mapper = dict(zip(sorted(values), RdYlBu11))
colors = [color_mapper[value] for value in values]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(
    categories=categories,
    values=values,
    colors=colors
))

# Create the figure
p = figure(x_range=categories, height=350, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Add the bars
p.vbar(x='categories', top='values', width=0.9, source=source, line_color='white',
       fill_color='colors')

# Customize the chart
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'

# Add hover tool
hover = HoverTool(tooltips=[('Category', '@categories'), ('Value', '@values')])
p.add_tools(hover)

# Show the plot
show(p)
```

##Q20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'.

##Ans.

```python
import plotly.graph_objects as go
import numpy as np

# Generate random data
np.random.seed(42)  # for reproducibility
x = np.linspace(0, 10, 100)
y = np.random.randn(100).cumsum()

# Create the figure
fig = go.Figure()

# Add the line trace
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Walk'))

# Update layout
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X Axis',
    yaxis_title='Y Axis',
    showlegend=True
)

# Show the plot
fig.show()
```

##Q21. Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set the title as 'Interactive Pie Chart'.

##Ans.

```python
import plotly.graph_objects as go
import random

# Generate random data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [random.randint(1, 100) for _ in categories]

# Calculate percentages
total = sum(values)
percentages = [f'{(value/total)*100:.1f}%' for value in values]

# Create labels with category names and percentages
labels = [f'{cat} ({pct})' for cat, pct in zip(categories, percentages)]

# Create the figure
fig = go.Figure(data=[go.Pie(
    labels=labels,
    values=values,
    textinfo='value',
    hoverinfo='label+percent',
    textfont_size=14,
    marker=dict(line=dict(color='#000000', width=2))
)])

# Update layout
fig.update_layout(
    title={
        'text': 'Interactive Pie Chart',
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    },
    font=dict(size=14)
)

# Show the plot
fig.show()
```
